# Distribution of Features

## Config

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

DATA_DIR = "../original_data"
TRAIN_VALUES_FILE = os.path.join(DATA_DIR, "train_values.csv")

sns.set_theme(style="whitegrid")
train_values = pd.read_csv(TRAIN_VALUES_FILE)
```

### geo_level_1_id

```python
axes = sns.countplot(x=train_values["geo_level_1_id"])
axes.xaxis.set_ticks(range(0, 31, 3))

plt.show()
```

We can see here that the distribution of samples to each geo level is unbalanced.

### count_floors_pre_eq

```python
sorted_values = train_values["count_floors_pre_eq"].value_counts().sort_index()
greater_than_three = sorted_values.loc[4:].sum()
res = sorted_values.loc[:3]
res[">= 4"] = greater_than_three
sns.barplot(x=res.index, y=res)

plt.show()
```

A majority of the samples have floors smaller than 4.


### age

```python
axes = sns.countplot(x=train_values["age"])
axes.xaxis.set_ticks(range(0, 41, 4))
axes.get_xticks()

plt.show()
```

Age feature seems like a skewed normal distribution with some outlier values after 200.

### area_percentage

```python
sorted_values = train_values["area_percentage"].value_counts().sort_index()
greater_than_or_equals_twenty_one = sorted_values.loc[21:].sum()
res = sorted_values.loc[:20]
res[">= 21"] = greater_than_or_equals_twenty_one
axes = sns.barplot(x=res.index, y=res);
axes.xaxis.set_ticks(list(range(0, 19)) + [20])

plt.show()
```

Area percentage also represents a skewed normal distribution.

### height_percentage

```python
sorted_values = train_values["height_percentage"].value_counts().sort_index()
greater_than_or_equals_twenty_one = sorted_values.loc[11:].sum()
res = sorted_values.loc[:10]
res[1] = 0
res = res.sort_index()
res
res[">= 11"] = greater_than_or_equals_twenty_one
axes = sns.barplot(x=res.index, y=res)

plt.show()
```

### has_superstructure_*

```python
secondaries = ('has_superstructure_adobe_mud',
                   'has_superstructure_mud_mortar_stone',
                   'has_superstructure_stone_flag',
                   'has_superstructure_cement_mortar_stone',
                   'has_superstructure_mud_mortar_brick',
                   'has_superstructure_cement_mortar_brick',
                   'has_superstructure_timber',
                   'has_superstructure_bamboo',
                   'has_superstructure_rc_non_engineered',
                   'has_superstructure_rc_engineered',
                   'has_superstructure_other')

fig, axes = plt.subplots(4, 3, figsize=(15, 15))
axes[-1, -1].axis("off")

labels = ["False", "True"]

for i, col in enumerate(secondaries):
    ax = axes[int(i/3), i%3]
    ax.set_title(col)
    ax.pie(x=train_values[col].value_counts().sort_index(), labels=labels, autopct="%.1f%%")

plt.show()
```

### count_families

```python
sorted_values = train_values["count_families"].value_counts().sort_index()
greater_than_or_equals_three = sorted_values.loc[3:].sum()
res = sorted_values.loc[:2]
res = res.sort_index()
res
res[">= 3"] = greater_than_or_equals_three
axes = sns.barplot(x=res.index, y=res)

plt.show()
```

Count families are mostly 1 with very few samples with a family size greater or equal than 3.
